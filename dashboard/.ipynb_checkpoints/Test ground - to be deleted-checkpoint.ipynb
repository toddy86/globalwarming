```python
import pymysql.cursors
import pandas as pd
```

```python
connection = pymysql.connect(host='localhost',
                             user='root',
                             password='',
                             db='globalwarming',
                             charset='utf8mb4',
                             cursorclass=pymysql.cursors.DictCursor)
```

```python
sql_query = """
SELECT 
    StationRecords.StationId as StationId, 
    StationRecords.Year as Year,
    StationRecords.`Month` as `Month`,
    StationRecords.AvgMaxTemp as AvgMaxTemp,
    StationRecords.AvgMinTemp as AvgMinTemp,
    StationRecords.MaxTemp as MaxTemp,
    StationRecords.MinTemp as MinTemp,
    StationRecords.MaxMinTemp as MaxMinTemp,
    CONCAT(StationDetails.City, ",", StationDetails.Country) as City
FROM StationRecords
JOIN StationDetails ON StationRecords.StationId = StationDetails.StationId
WHERE CONCAT(StationDetails.City, ",", StationDetails.Country) = 'Zermatt,Switzerland'
""" 
```

```python
def fahrenheit_to_celsius(df, columns):
    '''
    df = pd.DataFrame, containing one or more columns which are in degrees Fahrenheit
    columns = list of strings corresponding to the columns which are to be converted into degrees celsius 
    returns pd.Dataframe
    '''

    # For each specified column, convert the temperature from Fahrenheit to celsius
    for i in columns:
        df.loc[:,i] = (df.loc[:,i] - 32) * 5 / 9

    return df 
```

```python
df = pd.read_sql(sql_query, connection)
```

```python
test = df
test["AvgMaxTemp"] = (test.loc[:,"AvgMaxTemp"] - 32) * 5 / 9
```

```python
df.head()
df = df.groupby(['City','Year']).mean() 
df = df.reset_index()  
df.tail()
```

```python
def test_func(col):
    start_year = min(df.Year)
    end_year = max(df.Year)
    start_temp = df.loc[df["Year"] == start_year, col].unique()[0]
    end_temp = df.loc[df["Year"] == end_year, col].unique()[0]
    temp_change = end_temp - start_temp
    return temp_change
```
